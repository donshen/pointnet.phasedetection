```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
a = np.loadtxt('coord_O_hpl_1.pts')
```

```python
b = np.array([[ 1.55384362e+00,  6.97023571e-01,  -1.76950783e-1],
  [ 3.83118570e-1,  1.59434724e+00, -1.18837193e-1],
  [ 2.68495083e-1,  2.88120389e-1,  7.28940129e-1]])
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.scatter3D(a[:,0],a[:,1],a[:,2],c=np.mean(a,1),cmap='viridis',alpha = 0.5)
```

```python
plt.style.use('seaborn-poster')
rot = np.dot(a,b)

fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')


ax.scatter3D(rot[:,0],rot[:,1],rot[:,2],c=np.mean(rot,1), cmap='viridis',alpha = 0.5)
```
